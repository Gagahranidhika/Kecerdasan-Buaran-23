```python
# Mengimpor library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur agar plot ditampilkan di dalam notebook
%matplotlib inline

try:
    df = pd.read_csv('Employee Satisfaction Index.csv')
except FileNotFoundError:
    print("Error: File 'Employee Satisfaction Index.csv' tidak ditemukan.")
```

# 1. Menampilkan statistik deskriptif untuk atribut bertipe angka

```python
print("1. Statistik Deskriptif untuk Atribut Angka:")
# Memilih kolom numerik saja (termasuk 'satisfied' karena bernilai 0 atau 1)
numerical_attributes = df.select_dtypes(include=['number'])
print(numerical_attributes.describe())
```

# 2. Menampilkan 10 record pertama dari atribut bertipe angka

```python
print("\n2. 10 Record Pertama dari Atribut Angka:")
print(numerical_attributes.head(10))
```

# 3. Menampilkan informasi umum dataset

```python
print("\n3. Informasi Dataset:")
print(f"Jumlah Record: {df.shape[0]}")
print(f"Jumlah Atribut: {df.shape[1]}")
print(f"Nama Semua Atribut: {df.columns.tolist()}")
print(f"Jumlah Atribut dengan Tipe Angka: {len(numerical_attributes.columns)}")
```

# 4. Menampilkan jumlah label dari semua atribut yang bernilai object

```python
print("\n4. Jumlah Label dari Atribut Bertipe Objek:")
object_attributes = df.select_dtypes(include=['object'])
for col in object_attributes.columns:
    print(f"\n----- Atribut: {col} -----")
    print(df[col].value_counts())
```

# 5. Visualisasi dataset dengan grafik yang paling cocok

```python
print("\n5. Visualisasi: Tingkat Kepuasan Karyawan Berdasarkan Departemen")

plt.figure(figsize=(12, 7)) # Mengatur ukuran gambar
sns.countplot(data=df, x='Dept', hue='satisfied', palette='viridis')

# Menambahkan judul dan label
plt.title('Tingkat Kepuasan Karyawan Berdasarkan Departemen', fontsize=16)
plt.xlabel('Departemen', fontsize=12)
plt.ylabel('Jumlah Karyawan', fontsize=12)
plt.xticks(rotation=45, ha='right') # Memutar label x agar tidak tumpang tindih
plt.legend(title='Tingkat Kepuasan', labels=['Tidak Puas', 'Puas'])
plt.tight_layout() # Menyesuaikan plot agar pas

# Menampilkan plot
plt.show()
```

# 6. Menampilkan korelasi dataset menggunakan heatmap

```python
print("\n6. Visualisasi: Heatmap Korelasi Antar Atribut Angka")

plt.figure(figsize=(12, 10)) # Mengatur ukuran gambar

# Menghitung matriks korelasi dari atribut angka
# Kolom 'Unnamed: 0' dan 'emp_id' biasanya tidak relevan untuk korelasi
corr_matrix = numerical_attributes.corr()

# Membuat heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap Korelasi', fontsize=16)

# Menampilkan plot
plt.show()
```

Ini mengindikasikan bahwa semakin tinggi level pekerjaan seseorang, gajinya juga cenderung semakin tinggi. 3.  **Tingkat Kepuasan**: Secara umum, lebih dari separuh karyawan (**52.2%**) dalam dataset ini menyatakan puas. Visualisasi berdasarkan departemen menunjukkan tidak ada anomali yang signifikan; semua departemen memiliki proporsi karyawan puas dan tidak puas yang hampir sebanding.
