```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

%matplotlib inline

try:
    df = pd.read_csv('Employee Satisfaction Index.csv')
    # Menghapus kolom 'Unnamed: 0' yang merupakan duplikat dari index
    if 'Unnamed: 0' in df.columns:
        df = df.drop('Unnamed: 0', axis=1)
    print("Shape awal dataset:", df.shape)
except FileNotFoundError:
    print("Error: File 'Employee Satisfaction Index.csv' tidak ditemukan.")
```

```python
# 1.1 Cek missing values
print("Jumlah missing values per kolom:")
print(df.isnull().sum())
```

Tidak ada Missing Value

```python
# 1.2 Cek dan tangani data duplikat
duplicate_rows = df.duplicated().sum()
print(f"\nJumlah baris data duplikat: {duplicate_rows}")

if duplicate_rows > 0:
    df.drop_duplicates(inplace=True)
    print("Data duplikat telah dihapus.")
    print(f"Shape dataset setelah menghapus duplikat: {df.shape}")
```

Tidak ada data duplikat

```python
# 1.3 Deteksi outlier dengan Box Plot sebelum penanganan
numerical_cols_check = ['age', 'salary']
plt.figure(figsize=(15, 6))
for i, col in enumerate(numerical_cols_check):
    plt.subplot(1, 2, i+1)
    sns.boxplot(y=df[col], color='cyan')
    plt.title(f'Box Plot untuk {col} (Sebelum Penanganan Outlier)')
    plt.ylabel(col)
plt.tight_layout()
plt.show()

# Fungsi untuk menangani outlier dengan metode IQR Capping
def handle_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Capping: Ganti nilai outlier dengan batas atas/bawah
    dataframe[column] = np.where(dataframe[column] < lower_bound, lower_bound, dataframe[column])
    dataframe[column] = np.where(dataframe[column] > upper_bound, upper_bound, dataframe[column])
    
    return dataframe

# Terapkan fungsi pada kolom yang relevan
for col in numerical_cols_check:
    df = handle_outliers_iqr(df, col)

# Tampilkan Box Plot setelah penanganan outlier
plt.figure(figsize=(15, 6))
for i, col in enumerate(numerical_cols_check):
    plt.subplot(1, 2, i+1)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(f'Box Plot untuk {col} (Setelah Penanganan Outlier)')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

Tidak ada outlier pada kolom **age** dan **salary**. Hanya dilakukan pengecekan di ke 2 kolom itu karena 
sifat dari datanya yang memiliki **nilai range yang luas** dan **angka**.

```python
# 4. Feature Engineering: Membuat fitur baru 'tenure_per_age'
# Fitur ini merepresentasikan rasio antara level pekerjaan dengan usia.
# Nilai yang lebih tinggi bisa mengindikasikan promosi yang lebih cepat relatif terhadap usia.
df['tenure_per_age'] = df['job_level'] / df['age']

print("Fitur baru 'tenure_per_age' berhasil dibuat.")
print(df[['age', 'job_level', 'tenure_per_age']].head())
```

Feature **tenure_per_age dengan** dibuat dengan membagi j**ob_level** dengan **age**. Logikanya adalah fitur ini dapat menangkap efisiensi karier seorang karyawan. Seseorang dengan usia muda tetapi memiliki job_level tinggi akan memiliki rasio yang tinggi, yang mungkin berkorelasi dengan kepuasan atau ketidakpuasan kerja (misalnya, tekanan kerja yang lebih tinggi).

```python
# 5. Normalisasi/Standarisasi dan Encoding
X = df.drop(['satisfied', 'emp_id'], axis=1)
y = df['satisfied']

# Identifikasi kolom numerik dan kategorikal
numerical_features = X.select_dtypes(include=np.number).columns.tolist()
categorical_features = X.select_dtypes(exclude=np.number).columns.tolist()

print(f"\nKolom Numerik: {numerical_features}")
print(f"Kolom Kategorikal: {categorical_features}")

# Untuk numerik: StandardScaler
# Untuk kategorikal: OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# Terapkan preprocessor pada data X
X_processed = preprocessor.fit_transform(X)

print(f"Shape data setelah diproses: {X_processed.shape}")
```

**StandardScaler** digunakan untuk mengubah distribusi fitur numerik agar memiliki rata-rata 0 dan standar deviasi 1.

**OneHotEncoder** digunakan untuk kolom kategorikal. Encoder ini akan membuat kolom biner baru untuk setiap kategori unik di dalam sebuah fitur.

**ColumnTransformer** efisien untuk menerapkan transformasi yang berbeda pada kolom yang berbeda secara bersamaan.

```python
# 6. Splitting data
# Pembagian data: 80% untuk training, 20% untuk testing
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=42, stratify=y
)

print("\nDataset berhasil dibagi menjadi data training dan testing:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

**test_size=0.2** digunakan untuk mengalokasikan 20% dari data untuk pengujian dan 80% sisanya untuk pelatihan.

**random_state=42** digunakan untuk memastikan bahwa pembagian data akan selalu sama setiap kali kode ini dijalankan.

**stratify=y** adalah parameter untuk data klasifikasi. Ini memastikan bahwa proporsi nilai pada kolom target (satisfied 0 dan 1) akan sama di data training dan testing. Untuk mencegah kasus di mana salah satu set data secara kebetulan memiliki lebih banyak sampel dari satu kelas.

```python
# Mengimpor library yang dibutuhkan untuk model dan evaluasi
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

svc_model = SVC(kernel='rbf', random_state=42)
svc_model.fit(X_train, y_train)
```

### 1. Pelatihan Model Support Vector Classifier (SVC)

Pada langkah ini, dilakukan persiapan dan pelatihan model klasifikasi pertama, yaitu **Support Vector Classifier (SVC)**.

-   **SVC(kernel='rbf', random_state=42)**: Kode ini menginisialisasi atau membuat sebuah instance dari model SVC.
    -   `kernel='rbf'` merupakan pilihan yang fleksibel dan kuat untuk menemukan batas pemisah yang kompleks antar kelas.
    -   `random_state=42` digunakan untuk memastikan bahwa hasil dari model akan selalu konsisten setiap kali kode dijalankan.
-   **svc_model.fit(X_train, y_train)**: Ini adalah proses pelatihan model. Model akan menganalisis data `X_train` (fitur karyawan) dan `y_train` (label kepuasan) untuk menemukan pola terbaik yang bisa membedakan antara karyawan yang puas dan tidak puas.

```python
# Mengimpor model Random Forest
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
```

Selanjutnya, dilakukan pelatihan model kedua sebagai pembanding, yaitu **Random Forest Classifier**. Model ini bekerja dengan membangun banyak "pohon keputusan" dan mengambil suara mayoritas untuk hasil akhirnya.

-   **RandomForestClassifier(n_estimators=100, random_state=42)**: Kode ini membuat instance dari model Random Forest.
    -   `n_estimators=100` berarti model akan membangun 100 pohon keputusan yang berbeda untuk "berdiskusi" sebelum membuat prediksi.
-   **rf_model.fit(X_train, y_train)**: Ini adalah proses pelatihan di mana model mempelajari pola dari data training.
-   **y_pred_rf = rf_model.predict(X_test)**: Setelah dilatih, model digunakan untuk membuat tebakan (prediksi) pada data testing (`X_test`) yang belum pernah dilihatnya.

```python
# Membuat prediksi pada data testing
y_pred_svc = svc_model.predict(X_test)

print("\n--- Hasil Evaluasi Support Vector Classifier (SVC) ---")

# Menghitung akurasi
accuracy_svc = accuracy_score(y_test, y_pred_svc)
print(f"Akurasi Model SVC: {accuracy_svc:.2%}\n")

# Menampilkan Classification Report untuk metrik yang lebih detail
print("Classification Report untuk SVC:")
print(classification_report(y_test, y_pred_svc, target_names=['Tidak Puas (0)', 'Puas (1)']))

# Menampilkan Confusion Matrix secara visual
print("Confusion Matrix untuk SVC:")
cm_svc = confusion_matrix(y_test, y_pred_svc)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_svc, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Prediksi Tidak Puas', 'Prediksi Puas'],
            yticklabels=['Aktual Tidak Puas', 'Aktual Puas'])
plt.title('Confusion Matrix - SVC', fontsize=16)
plt.ylabel('Label Aktual')
plt.xlabel('Label Prediksi')
plt.show()
```
